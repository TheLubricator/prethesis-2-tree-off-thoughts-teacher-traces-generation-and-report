````python
# ============================================================================
# VISUALIZATION & ANALYSIS HELPER FUNCTIONS
# ============================================================================

def visualize_tree_codeact(root_node, max_depth=3):
    """
    Visualize the CodeAct search tree structure
    
    Args:
        root_node: Root TreeNode of the search tree
        max_depth: Maximum depth to display
    """
    def print_node(node, depth=0, prefix="", is_last=True):
        if depth > max_depth:
            return
        
        # Create the branch characters
        branch = "└── " if is_last else "├── "
        extension = "    " if is_last else "│   "
        
        # Format node information
        if node.depth == 0:
            line = f"🌳 ROOT: {node.state or 'Start'}"
        else:
            # Get action summary
            action = node.action[:50] + "..." if len(node.action) > 50 else node.action
            value_str = f"(value: {node.value:.2f})" if node.value > 0 else ""
            solution_marker = " ✅ SOLUTION" if node.is_solution else ""
            pruned_marker = " ❌ PRUNED" if node.is_pruned else ""
            
            line = f"{branch}[D{node.depth}] {action} {value_str}{solution_marker}{pruned_marker}"
        
        print(prefix + line)
        
        # Print children
        children = list(node.children.values())
        for i, child in enumerate(children):
            is_last_child = (i == len(children) - 1)
            new_prefix = prefix + extension
            print_node(child, depth + 1, new_prefix, is_last_child)
    
    print("\n" + "="*70)
    print("🎄 SEARCH TREE VISUALIZATION")
    print("="*70)
    print_node(root_node)
    print("="*70 + "\n")


def analyze_tree(root_node):
    """
    Analyze and display statistics about the search tree
    
    Args:
        root_node: Root TreeNode of the search tree
    """
    stats = {
        'total_nodes': 0,
        'by_depth': {},
        'solutions': 0,
        'pruned': 0,
        'max_depth': 0,
        'avg_branching': 0,
        'total_children': 0
    }
    
    def traverse(node):
        stats['total_nodes'] += 1
        depth = node.depth
        
        # Track depth statistics
        if depth not in stats['by_depth']:
            stats['by_depth'][depth] = 0
        stats['by_depth'][depth] += 1
        
        # Track max depth
        if depth > stats['max_depth']:
            stats['max_depth'] = depth
        
        # Track solutions and pruned nodes
        if node.is_solution:
            stats['solutions'] += 1
        if node.is_pruned:
            stats['pruned'] += 1
        
        # Track branching factor
        num_children = len(node.children)
        if num_children > 0:
            stats['total_children'] += num_children
        
        # Recurse to children
        for child in node.children.values():
            traverse(child)
    
    # Collect statistics
    traverse(root_node)
    
    # Calculate average branching factor
    internal_nodes = stats['total_nodes'] - stats['by_depth'].get(stats['max_depth'], 0)
    stats['avg_branching'] = stats['total_children'] / internal_nodes if internal_nodes > 0 else 0
    
    # Display statistics
    print("\n" + "="*70)
    print("📊 TREE ANALYSIS")
    print("="*70)
    print(f"Total Nodes:        {stats['total_nodes']}")
    print(f"Max Depth:          {stats['max_depth']}")
    print(f"Solutions Found:    {stats['solutions']}")
    print(f"Pruned Nodes:       {stats['pruned']}")
    print(f"Avg Branching:      {stats['avg_branching']:.2f}")
    print()
    print("Nodes by Depth:")
    for depth in sorted(stats['by_depth'].keys()):
        count = stats['by_depth'][depth]
        bar = "█" * min(count, 50)
        print(f"  Depth {depth}: {count:3d} {bar}")
    print("="*70 + "\n")


def export_solution_steps(solution_node, filename="solution.txt"):
    """
    Export the solution path to a text file
    
    Args:
        solution_node: TreeNode representing a solution
        filename: Output filename
    """
    if not solution_node.is_solution:
        print(f"⚠️ Warning: Node {solution_node.id} is not marked as a solution")
    
    # Collect path from root to solution
    path = []
    current = solution_node
    while current is not None:
        path.append(current)
        current = current.parent
    
    path.reverse()  # Start from root
    
    # Write to file
    with open(filename, 'w', encoding='utf-8') as f:
        f.write("="*70 + "\n")
        f.write("GAME OF 24 - SOLUTION PATH\n")
        f.write("="*70 + "\n\n")
        
        for i, node in enumerate(path):
            if node.depth == 0:
                f.write(f"START: {node.state}\n")
            else:
                f.write(f"\nStep {node.depth}:\n")
                f.write(f"  Action: {node.action}\n")
                f.write(f"  Result: {node.state}\n")
                f.write(f"  Value:  {node.value:.3f}\n")
                
                # Write CodeAct details if available
                if hasattr(node, 'codeact') and node.codeact:
                    f.write(f"\n  CodeAct Details:\n")
                    f.write(f"    Thought: {node.codeact.get('thought', 'N/A')}\n")
                    f.write(f"    Code: {node.codeact.get('code', 'N/A')}\n")
                    f.write(f"    Observation: {node.codeact.get('observation', 'N/A')}\n")
        
        f.write("\n" + "="*70 + "\n")
        if solution_node.is_solution:
            f.write("✅ SOLUTION VERIFIED\n")
        f.write("="*70 + "\n")
    
    print(f"✅ Solution exported to: {filename}")


print("✅ Visualization and analysis functions loaded!")
print("   - visualize_tree_codeact(root, max_depth=3)")
print("   - analyze_tree(root)")
print("   - export_solution_steps(solution_node, filename)")
````

````python
# ============================================================================
# SAFE SANDBOX FOR CODE EXECUTION
# ============================================================================

import sys
from io import StringIO
import ast
import re

class SafeAgentSandbox:
    """A SECURE sandbox that only allows safe math operations"""
    def __init__(self):
        # ONLY allow these safe built-ins
        self.safe_builtins = {
            'print': print,
            'range': range,
            'len': len,
            'sum': sum,
            'min': min,
            'max': max,
            'abs': abs,
            'round': round,
            'int': int,
            'float': float,
            'str': str,
            'list': list,
        }
        self.globals = {
            "__builtins__": self.safe_builtins,
            "math": __import__("math"),
        }
        
    def run(self, code):
        """Execute code in restricted environment with validation"""
        # Blacklist dangerous operations
        dangerous_keywords = ['import', 'open', 'exec', 'eval', '__', 'os', 'sys', 'subprocess', 'file']
        if any(keyword in code.lower() for keyword in dangerous_keywords):
            return "Error: Code contains forbidden operations", None
        
        # CRITICAL FIX: Check for number reuse (e.g., numbers[1] + numbers[1])
        # This prevents invalid solutions like "6 + 6 = 12" from [20, 6]
        index_pattern = r'numbers\[(\d+)\]'
        indices_used = re.findall(index_pattern, code)
        if len(indices_used) != len(set(indices_used)):
            return "Error: Cannot use the same number twice in one operation", None
        
        # NEW FIX: Prevent using arbitrary numbers not in the 'numbers' array
        # This prevents invalid moves like "2.4 * 10" when only [2.4, 5] is available
        lines = code.split('\n')
        for line in lines:
            # Skip comments
            if '#' in line:
                line = line.split('#')[0]
            line = line.strip()
            
            # Skip empty lines and variable definitions
            if not line or line.startswith('numbers = ') or line.startswith('remaining = ') or line.startswith('new_numbers = ') or line.startswith('print('):
                continue
                
            # Check if this is an arithmetic operation line (res = ...)
            if line.startswith('res = ') and any(op in line for op in [' + ', ' - ', ' * ', ' / ', ' % ', ' // ']):
                # Extract the right side of the assignment
                rhs = line.split('res = ', 1)[1]
                
                # Find all numeric literals (but not inside brackets)
                # Pattern: number not preceded by '[' and not followed by ']'
                numeric_literals = re.findall(r'(?<!\[)\b(\d+(?:\.\d+)?)\b(?!\])', rhs)
                
                for num_str in numeric_literals:
                    # Check if this appears as an array index
                    if f'[{num_str}]' in rhs:
                        continue  # This is an array index, OK
                    
                    # This is a bare numeric literal in arithmetic - NOT ALLOWED
                    return f"Error: Cannot use arbitrary number {num_str} - must use only numbers from the array via numbers[index]", None
        
        try:
            # Capture standard output
            output_buffer = StringIO()
            sys.stdout = output_buffer
            
            # Execute with restricted globals
            exec(code, self.globals)
            
            sys.stdout = sys.__stdout__
            output = output_buffer.getvalue().strip() or "Code executed successfully."
            
            # Return output and the updated 'numbers' variable if it exists
            new_state = self.globals.get('new_numbers', None)
            
            # Additional validation: check if result uses exactly 2 numbers from input
            original_count = len(self.globals.get('numbers', []))
            new_count = len(new_state) if new_state else 0
            expected_count = original_count - 1  # Combine 2 numbers into 1
            
            if new_state and new_count != expected_count:
                return f"Error: Invalid operation - expected {expected_count} numbers, got {new_count}", None
            
            return output, new_state
            
        except Exception as e:
            sys.stdout = sys.__stdout__
            return f"Error: {str(e)}", None

# Initialize global sandbox
sandbox = SafeAgentSandbox()

# Test the sandbox
print("✓ Safe Sandbox initialized")
test_code = """
numbers = [4, 5, 6, 10]
res = numbers[0] + numbers[1]
remaining = numbers[2:]
new_numbers = [res] + remaining
print(new_numbers)
"""
test_output, test_state = sandbox.run(test_code)
print(f"Test: {test_output} → {test_state}")
````

## 🔧 Delayed Fraction Preservation (DFP)

**The Problem:** Gemini kills fragile fractional states too early, even when they're on the solution path.

**Example Failure:**
- Puzzle: [3,9,9,13] → Solution requires 13-9=4, then 9/(4-3)=9, then 9+9=18... (actually 9+9-4+3 won't work)
- Puzzle: [4,7,8,10] → Solution requires 4/7=0.571, then 8/0.571=14, then 10+14=24
- These fractional intermediate states get pruned too early!

**The Fix:** If a state has exactly 1 non-integer + at least 1 small integer (≤6), give it **one more depth** before pruning.

**Why it works:** Hard Game of 24 puzzles follow this pattern:
1. Create fragile fraction (e.g., 4/7 = 0.571)
2. Preserve it (don't prune immediately)
3. Combine at the end (e.g., 8/0.571 = 14, then 10+14 = 24)

This single bias prevents premature pruning of promising paths without changing prompts!

````python
# 🔧 Delayed Fraction Preservation (DFP) - Helper Function
def state_signature(nums: List[float]) -> tuple:
    """
    Classify state for Delayed Fraction Preservation (DFP).
    
    Returns:
        (non_int_count, small_int_count) tuple
        - non_int_count: How many non-integers in state
        - small_int_count: How many small integers (≤6) in state
    
    DFP Rule: If non_int_count == 1 AND small_int_count >= 1,
              this state should be preserved for one more depth.
    """
    non_int = [x for x in nums if abs(x - round(x)) > 1e-6]
    small_int = [x for x in nums if abs(x - round(x)) < 1e-6 and x <= 6]
    return len(non_int), len(small_int)

print("✓ DFP helper function defined")
````

````python
# Helper Functions
def get_current_numbers(state: str) -> str:
    """Extract the remaining numbers from the current state"""
    if not state.strip():
        return ""
    last_line = state.strip().split('\n')[-1]
    if 'left:' in last_line:
        return last_line.split('left: ')[-1].split(')')[0]
    return ""

def validate_solution(numbers: List[int], solution: str) -> bool:
    """
    Validate if a solution is correct
    
    Args:
        numbers: Original input numbers
        solution: The proposed solution string
    
    Returns:
        True if solution is valid and equals 24
    """
    try:
        # Extract the final expression
        expression = solution.strip().split('\n')[-1].lower()
        expression = expression.replace('answer: ', '').split('=')[0]
        
        # Extract numbers used in the expression
        used_numbers = re.findall(r'\d+', expression)
        
        # Check if same numbers are used
        if sorted(used_numbers) != sorted([str(n) for n in numbers]):
            return False
        
        # Evaluate the expression
        result = sympy.simplify(expression)
        return int(result) == 24
    except Exception as e:
        print(f"Validation error: {e}")
        return False

def format_numbers(numbers: List[int]) -> str:
    """Format numbers as space-separated string"""
    return ' '.join(map(str, numbers))

print("✓ Helper functions defined")
````

````python
# Tree Node Class for ToT with CodeAct
class TreeNode:
    """Represents a node in the Tree of Thoughts with CodeAct support"""
    
    node_counter = 0  # Global counter for unique node IDs
    
    def __init__(self, state: str, parent=None, action: str = "", value: float = 0.0,
                 thought: str = "", code: str = "", observation: str = "", 
                 path_history: str = ""):
        """
        Initialize a tree node
        
        Args:
            state: Current state (accumulated steps)
            parent: Parent TreeNode
            action: Action taken to reach this state (for display)
            value: Evaluation value of this state
            thought: The reasoning/thought process (CodeAct)
            code: The executable Python code (CodeAct)
            observation: The execution result (CodeAct)
            path_history: Full history of steps from root (like original ToT paper)
        """
        TreeNode.node_counter += 1
        self.id = TreeNode.node_counter
        self.state = state
        self.parent = parent
        self.action = action
        self.value = value
        self.children = []
        self.depth = 0 if parent is None else parent.depth + 1
        self.is_solution = False
        self.is_pruned = False
        
        # CodeAct specific attributes
        self.thought = thought
        self.code = code
        self.observation = observation
        
        # Path history - like original ToT paper
        # This accumulates all steps from root to current node
        self.path_history = path_history
        
        # Evaluation record for distillation (added for SLM training)
        self.evaluation = None  # Will be populated during evaluate_state()
        
        if parent:
            parent.children.append(self)
    
    def to_dict(self) -> Dict[str, Any]:
        """Convert node to dictionary for JSON serialization"""
        return {
            'id': self.id,
            'state': self.state,
            'action': self.action,
            'value': float(self.value),
            'depth': self.depth,
            'is_solution': self.is_solution,
            'is_pruned': self.is_pruned,
            'parent_id': self.parent.id if self.parent else None,
            'children_ids': [child.id for child in self.children],
            'current_numbers': get_current_numbers(self.state) if self.state else "",
            # CodeAct information
            'codeact': {
                'thought': self.thought,
                'code': self.code,
                'observation': self.observation
            },
            # Path history (like original ToT)
            'path_history': self.path_history,
            # Evaluation record (for distillation dataset)
            'evaluation': self.evaluation if self.evaluation else None
        }
    
    def __repr__(self):
        return f"Node(id={self.id}, depth={self.depth}, value={self.value:.3f}, children={len(self.children)})"

# Reset counter for fresh start
TreeNode.node_counter = 0
print("✓ TreeNode class with CodeAct support and path history defined")
````

````python
# Game of 24 Tree of Thoughts Solver with CodeAct
import time
from datetime import datetime

class Game24TreeOfThoughts:
    """Tree of Thoughts solver for Game of 24 using CodeAct pattern"""
    
    def __init__(self, 
                 temperature: float = 0.7,
                 n_evaluate_sample: int = 3,
                 n_select_sample: int = 15,
                 max_steps: int = 6,
                 api_delay: float = 1.0,
                 selection_method: str = 'greedy',
                 exhaustive_depth1: bool = False,
                 enable_ser: bool = False):
        """
        Initialize the ToT solver with CodeAct
        
        Args:
            temperature: Sampling temperature for generation
            n_evaluate_sample: Number of evaluation samples per state
            n_select_sample: Beam width (how many states to keep per step) - increased to 15 for harder puzzles
            max_steps: Maximum number of steps (6 for harder puzzles, was 4)
            api_delay: Delay between API calls (rate limiting)
            selection_method: 'greedy' (top-k) or 'sample' (probability-based)
            exhaustive_depth1: If True, try ALL possible first moves (~24 operations) instead of LLM proposals
            enable_ser: If True, enable Selective Exhaustive Rescue when LLM proposals are weak
        """
        self.temperature = temperature
        self.n_evaluate_sample = n_evaluate_sample
        self.n_select_sample = n_select_sample
        self.max_steps = max_steps
        self.api_delay = api_delay
        self.selection_method = selection_method  # NEW: greedy or sample
        self.exhaustive_depth1 = exhaustive_depth1  # NEW: exhaustive first move mode
        self.enable_ser = enable_ser  # NEW: Selective Exhaustive Rescue toggle
        self.value_cache = {}  # Cache evaluations
        
        # Tree structure
        self.root = None
        self.all_nodes = []
        self.solutions = []
        
        # Statistics with daily tracking
        self.stats = {
            'total_nodes': 0,
            'api_calls': 0,
            'cache_hits': 0,
            'solutions_found': 0,
            'code_executions': 0,
            'code_errors': 0,
            'daily_requests': 0,  # Track requests per day
            'session_start': datetime.now()
        }
        
        # Free tier limits
        self.DAILY_LIMIT = 14000  # 14k requests per day
        self.MINUTE_LIMIT = 20    # 20 requests per minute
        
        print(f"✓ Solver initialized with rate limiting:")
        print(f"  • Delay: {api_delay}s between calls (~{int(60/api_delay)} req/min)")
        print(f"  • Daily limit: {self.DAILY_LIMIT:,} requests")
        print(f"  • Safety margin: {API_DELAY}s = ~17 req/min (vs 20 limit)")
        if exhaustive_depth1:
            print(f"  • Mode: EXHAUSTIVE DEPTH-1 (will try all ~24 possible first moves)")
        else:
            print(f"  • Mode: LLM PROPOSALS (standard beam search)")
        if enable_ser:
            print(f"  • SER: ENABLED (will rescue weak LLM proposals with exhaustive search)")
        else:
            print(f"  • SER: DISABLED (pure LLM-guided search)")
    
    def check_rate_limits(self):
        """Check if we're approaching rate limits"""
        # Check daily limit
        if self.stats['daily_requests'] >= self.DAILY_LIMIT * 0.9:  # 90% threshold
            print(f"⚠ WARNING: Approaching daily limit ({self.stats['daily_requests']}/{self.DAILY_LIMIT})")
            if self.stats['daily_requests'] >= self.DAILY_LIMIT:
                raise Exception(f"Daily API limit reached ({self.DAILY_LIMIT} requests)")
        
        # Log progress every 50 requests
        if self.stats['api_calls'] % 50 == 0 and self.stats['api_calls'] > 0:
            elapsed = (datetime.now() - self.stats['session_start']).total_seconds()
            rate = self.stats['api_calls'] / (elapsed / 60) if elapsed > 0 else 0
            print(f"📊 Progress: {self.stats['api_calls']} API calls | Rate: {rate:.1f} req/min")
    
    def execute_codeact_step(self, current_numbers: List[int], avoid_states: List[List] = None, use_high_temp: bool = False) -> Dict[str, Any]:
        """
        Execute one CodeAct step: Thought → Code → Observation
        
        Args:
            current_numbers: Current list of numbers
            avoid_states: List of states to avoid (already generated)
            use_high_temp: Use higher temperature for more diversity
        
        Returns:
            Dict with thought, code, observation, new_state, action
        """
        # Reset sandbox state
        sandbox.globals['numbers'] = list(current_numbers)
        
        # Create prompt
        remaining = len(current_numbers) - 1
        
        # Build avoid_states message
        avoid_msg = ""
        if avoid_states:
            avoid_msg = f"\nIMPORTANT: DO NOT generate these states (already tried):\n"
            for state in avoid_states[:5]:  # Show max 5 to avoid token bloat
                avoid_msg += f"  - {state}\n"
            avoid_msg += "Generate a DIFFERENT operation (try subtraction, division, or addition instead of multiplication)!\n"
        
        # Use improved prompts if enabled
        prompt_template = PROPOSE_PROMPT_CODEACT_IMPROVED if USE_IMPROVED_PROMPTS else PROPOSE_PROMPT_CODEACT
        
        prompt = prompt_template.format(
            input=current_numbers,
            remaining=remaining,
            avoid_states=avoid_msg
        )
        
        # Check rate limits before API call
        self.check_rate_limits()
        
        # Call API with rate limiting
        time.sleep(self.api_delay)
        self.stats['api_calls'] += 1
        self.stats['daily_requests'] += 1
        
        # Use higher temperature for diversity when generating multiple proposals
        temp = 1.4 if use_high_temp else self.temperature
        
        try:
            response = openai_generate(prompt, n=1, temperature=temp, system_prompt=SYSTEM_PROMPT)[0]
            
            # Parse response
            thought_match = re.search(r"Thought:\s*(.+?)(?=Code:|```)", response, re.DOTALL)
            code_match = re.search(r"```python\n(.*?)\n```", response, re.DOTALL)
            
            if not code_match:
                return {
                    'error': 'No code block found in response',
                    'thought': response[:100],
                    'code': '',
                    'observation': 'Error: No executable code generated',
                    'new_state': None,
                    'action': ''
                }
            
            thought = thought_match.group(1).strip() if thought_match else "No thought provided"
            code = code_match.group(1)
            
            # Execute code in sandbox
            self.stats['code_executions'] += 1
            sandbox.globals['numbers'] = list(current_numbers)
            observation, new_state = sandbox.run(code)
            
            # Parse new_state from observation if not available
            if new_state is None and observation and not observation.startswith("Error:"):
                try:
                    new_state = ast.literal_eval(observation.strip())
                except:
                    self.stats['code_errors'] += 1
                    return {
                        'error': 'Failed to parse new state',
                        'thought': thought,
                        'code': code,
                        'observation': observation,
                        'new_state': None,
                        'action': ''
                    }
            
            if observation.startswith("Error:"):
                self.stats['code_errors'] += 1
            
            # Create action string for display
            action = f"{observation}" if new_state else "Error in execution"
            
            return {
                'thought': thought,
                'code': code,
                'observation': observation,
                'new_state': new_state,
                'action': action
            }
            
        except Exception as e:
            self.stats['code_errors'] += 1
            return {
                'error': str(e),
                'thought': '',
                'code': '',
                'observation': f'Error: {str(e)}',
                'new_state': None,
                'action': ''
            }
    
    def get_proposals(self, current_numbers: List[int], original_input: List[int], path_history: str = "", n_proposals: int = 1, avoid_states: List[List] = None) -> List[Dict]:
        """
        Generate multiple proposal steps using CodeAct (following original ToT paper approach)
        
        Args:
            current_numbers: Current list of numbers
            original_input: Original puzzle numbers (for context)
            path_history: Accumulated history of steps taken so far
            n_proposals: How many proposals to generate
            avoid_states: List of states to avoid (already generated)
        
        Returns:
            List of proposal dicts (unique states only)
        """
        proposals = []
        seen_states = set()
        
        # Add avoid_states to seen
        if avoid_states:
            for state in avoid_states:
                seen_states.add(tuple(sorted(state)))
        
        # Build prompt with full context (original input + history + current state)
        # Use improved prompts if enabled
        prompt_template = PROPOSE_PROMPT_CODEACT_IMPROVED if USE_IMPROVED_PROMPTS else PROPOSE_PROMPT_CODEACT
        
        prompt = prompt_template.format(
            original_input=original_input,
            history=path_history if path_history else "(Starting state)",
            input=current_numbers,
            avoid_states=""  # For improved prompt compatibility
        )
        
        # Check rate limits
        self.check_rate_limits()
        time.sleep(self.api_delay)
        self.stats['api_calls'] += 1
        self.stats['daily_requests'] += 1
        
        # Generate ONE response with MULTIPLE proposals (like original ToT)
        try:
            response = openai_generate(prompt, n=1, temperature=1.0, system_prompt=SYSTEM_PROMPT)[0]
        except Exception as e:
            print(f"⚠ Error generating proposals: {e}")
            return []
        
        # Parse response - look for multiple proposals with Thought, Math, Remaining, Code
        # New format: Thought: ... \n Math: ... \n Remaining: ... \n Code: ```python ... ```
        full_pattern = r"Thought:\s*(.+?)\s*Math:\s*(.+?)\s*Remaining:\s*(.+?)(?:Code:.*?```python\n(.*?)\n```)"
        matches = re.findall(full_pattern, response, re.DOTALL | re.IGNORECASE)
        
        if not matches:
            # Fallback: Old format without Math/Remaining
            thought_code_pattern = r"Thought:\s*(.+?)(?:Code:.*?```python\n(.*?)\n```)"
            old_matches = re.findall(thought_code_pattern, response, re.DOTALL | re.IGNORECASE)
            # Convert to new format (thought, math, remaining, code)
            matches = [(m[0], "", "", m[1]) for m in old_matches]
        
        # Process each parsed step
        for match in matches:
            if len(proposals) >= n_proposals:
                break
            
            # Extract all fields (thought, math, remaining, code)
            thought = match[0].strip()
            math_calc = match[1].strip() if len(match) > 1 else ""
            remaining = match[2].strip() if len(match) > 2 else ""
            code = match[3].strip() if len(match) > 3 else ""
            
            if not code:
                continue
            
            # Enhance thought with mental calculation info
            if math_calc:
                thought = f"{thought} [{math_calc}]"
            if remaining:
                thought = f"{thought} → {remaining}"
            
            try:
                # Execute code in sandbox
                self.stats['code_executions'] += 1
                sandbox.globals['numbers'] = list(current_numbers)
                observation, new_state = sandbox.run(code)
                
                # Check if valid and unique
                if new_state and not observation.startswith("Error:"):
                    state_tuple = tuple(sorted(new_state))
                    
                    if state_tuple not in seen_states:
                        proposals.append({
                            'thought': thought,
                            'code': code,
                            'observation': observation,
                            'new_state': new_state,
                            'action': observation
                        })
                        seen_states.add(state_tuple)
            except Exception as e:
                continue
        
        return proposals
    
    def generate_all_first_moves(self, numbers: List[float]) -> List[Dict]:
        """
        Generate ALL possible first moves exhaustively (no LLM).
        
        For 4 numbers, there are:
        - C(4,2) = 6 pairs of numbers
        - 4 operations (+, -, *, /) per pair
        - 2 orders for non-commutative ops (a-b vs b-a, a/b vs b/a)
        - Total: ~24-30 distinct first moves
        
        Args:
            numbers: List of 4 numbers [a, b, c, d]
        
        Returns:
            List of proposal dicts with thought, code, observation, new_state
        """
        from itertools import combinations
        import math
        
        proposals = []
        operations = [
            ('+', lambda a, b: a + b, "{} + {} = {}"),
            ('-', lambda a, b: a - b, "{} - {} = {}"),
            ('*', lambda a, b: a * b, "{} × {} = {}"),
            ('/', lambda a, b: a / b if b != 0 else None, "{} ÷ {} = {}")
        ]
        
        print(f"  🔬 Exhaustively generating all first moves from {numbers}...")
        
        # Generate all pairs
        for i, j in combinations(range(len(numbers)), 2):
            a, b = numbers[i], numbers[j]
            remaining_indices = [k for k in range(len(numbers)) if k not in [i, j]]
            remaining = [numbers[k] for k in remaining_indices]
            
            for op_symbol, op_func, math_template in operations:
                # Try both orders for non-commutative operations
                for num1, num2, idx1, idx2 in [(a, b, i, j), (b, a, j, i)]:
                    # Skip redundant order for commutative operations
                    if op_symbol in ['+', '*'] and (num1, num2) != (a, b):
                        continue
                    
                    # Calculate result
                    result = op_func(num1, num2)
                    
                    # Skip invalid results
                    if result is None or not math.isfinite(result):
                        continue
                    
                    # Create new state
                    new_state = [result] + remaining
                    
                    # Create thought
                    math_str = math_template.format(num1, num2, result)
                    thought = f"Exhaustive: {math_str}"
                    
                    # Create code (matching the format expected by sandbox)
                    code = f"""numbers = {numbers}
res = numbers[{idx1}] {op_symbol} numbers[{idx2}]
remaining = {remaining}
new_numbers = [res] + remaining
print(new_numbers)"""
                    
                    # Create proposal
                    proposal = {
                        'thought': thought,
                        'code': code,
                        'observation': str(new_state),
                        'new_state': new_state,
                        'action': str(new_state)
                    }
                    
                    proposals.append(proposal)
        
        print(f"  ✓ Generated {len(proposals)} exhaustive first moves (no LLM calls!)")
        return proposals
    
    def evaluate_state(self, numbers: List[int], is_final: bool = False, 
                      original_numbers: List[int] = None) -> Tuple[float, dict]:
        """
        Evaluate how promising a state is (HYBRID: Heuristics + LLM)
        Returns detailed evaluation record for distillation dataset
        
        Args:
            numbers: Current numbers to evaluate
            is_final: Whether this is a final answer
            original_numbers: Original input numbers (for final validation)
        
        Returns:
            Tuple of (numeric_value, evaluation_record)
            - numeric_value: float (higher is better)
            - evaluation_record: dict with detailed reasoning for training SLMs
        """
        # Initialize evaluation record for distillation
        eval_record = {
            "state": str(numbers),
            "is_final": is_final,
            "heuristic_checks": {},
            "llm_judgments": [],
            "reasoning": [],
            "score_breakdown": {},
            "final_value": 0.0
        }
        
        # === HEURISTIC CHECKS (Fast, no LLM needed) ===
        
        # 1. Check if this is the final answer
        if len(numbers) == 1:
            eval_record["heuristic_checks"]["is_single_number"] = True
            if abs(numbers[0] - 24) < 0.001:  # Account for floating point
                eval_record["heuristic_checks"]["is_solution"] = True
                eval_record["reasoning"].append(f"Final state equals 24 - SOLUTION FOUND!")
                eval_record["final_value"] = 100.0
                return 100.0, eval_record  # Perfect solution!
            else:
                eval_record["heuristic_checks"]["is_solution"] = False
                eval_record["reasoning"].append(f"Final state is {numbers[0]}, not 24 - WRONG ANSWER")
                eval_record["final_value"] = 0.001
                return 0.001, eval_record  # Wrong answer
        
        # 2. CRITICAL: Penalize "premature 24" (the 4×6 trap!)
        # If 24 appears but we still have multiple numbers, this is BAD
        if 24 in numbers or 24.0 in numbers:
            eval_record["heuristic_checks"]["has_premature_24"] = True
            eval_record["reasoning"].append("Contains 24 but not final state - this is a dead-end trap!")
            eval_record["final_value"] = 0.01
            return 0.01, eval_record  # Very low score - this path is likely a dead end
        else:
            eval_record["heuristic_checks"]["has_premature_24"] = False
        
        # 3. Penalize EXTREMELY large numbers (>1000)
        # But allow moderate values like 30, 48, etc. which can lead to solutions
        # Example: [30, 4, 10] → 30-6=24 where 6 comes from 10-4
        max_abs = max(abs(n) for n in numbers)
        eval_record["heuristic_checks"]["max_abs_value"] = max_abs
        if max_abs > 1000:
            eval_record["heuristic_checks"]["has_huge_numbers"] = True
            eval_record["reasoning"].append(f"Very large number detected ({max_abs}) - hard to reduce to 24")
            eval_record["final_value"] = 0.1
            return 0.1, eval_record  # Low score for huge numbers
        else:
            eval_record["heuristic_checks"]["has_huge_numbers"] = False
        
        # 4. Penalize numbers that are all very small (< 1)
        # Hard to reach 24 from very small fractions
        if all(abs(n) < 0.5 for n in numbers):
            eval_record["heuristic_checks"]["all_tiny_numbers"] = True
            eval_record["reasoning"].append("All numbers are very small (<0.5) - hard to reach 24 from fractions")
            eval_record["final_value"] = 0.5
            return 0.5, eval_record  # Low score for tiny numbers
        else:
            eval_record["heuristic_checks"]["all_tiny_numbers"] = False
        
        # 5. BONUS: Reward states with numbers in "24-reachable range"
        # Having a number close to 24 (20-30) is often good!
        # Example: [30, 4, 10] has 30 which can become 24 via 30-6
        has_close_number = any(20 <= abs(n) <= 40 for n in numbers)
        has_small_number = any(1 <= abs(n) <= 10 for n in numbers)
        
        eval_record["heuristic_checks"]["has_number_near_24"] = has_close_number
        eval_record["heuristic_checks"]["has_small_adjusters"] = has_small_number
        
        # If we have one number near 24 AND small numbers to adjust it, that's promising!
        if has_close_number and has_small_number and len(numbers) >= 2:
            # Don't override LLM, but give a slight boost later
            llm_boost = 1.2  # Will multiply LLM score by this
            eval_record["heuristic_checks"]["promising_pattern"] = True
            close_nums = [n for n in numbers if 20 <= abs(n) <= 40]
            small_nums = [n for n in numbers if 1 <= abs(n) <= 10]
            eval_record["reasoning"].append(f"Promising pattern: has number(s) near 24 {close_nums} with small adjuster(s) {small_nums}")
        else:
            llm_boost = 1.0
            eval_record["heuristic_checks"]["promising_pattern"] = False
        
        eval_record["heuristic_checks"]["llm_boost_factor"] = llm_boost
        
        # === HARD-CODED 2-NUMBER CHECK (Bypass unreliable LLM arithmetic) ===
        # The LLM fails to recognize simple 2-number solutions like [15, 9] → 24
        # So we hard-code this check in Python for reliability
        if len(numbers) == 2:
            a, b = numbers[0], numbers[1]
            operations_to_24 = [
                abs(a + b - 24) < 0.001,      # a + b = 24
                abs(a - b - 24) < 0.001,      # a - b = 24
                abs(b - a - 24) < 0.001,      # b - a = 24
                abs(a * b - 24) < 0.001,      # a * b = 24
                abs(a / b - 24) < 0.001 if b != 0 else False,  # a / b = 24
                abs(b / a - 24) < 0.001 if a != 0 else False   # b / a = 24
            ]
            
            if any(operations_to_24):
                eval_record["heuristic_checks"]["two_number_reaches_24"] = True
                eval_record["reasoning"].append(f"Hard-coded check: {numbers} can reach 24 with ONE operation!")
                eval_record["final_value"] = 60.0  # Same as "sure, sure, sure"
                eval_record["llm_judgments"] = ["sure", "sure", "sure"]  # Fake LLM response
                eval_record["from_cache"] = False
                eval_record["score_breakdown"] = {
                    "sure_count": 3,
                    "likely_count": 0,
                    "impossible_count": 0,
                    "n_evaluate_sample": 3,
                    "raw_score": 60.0,
                    "boosted_score": 60.0,
                    "confidence_level": "high"
                }
                return 60.0, eval_record  # Guaranteed "sure" score!
            else:
                eval_record["heuristic_checks"]["two_number_reaches_24"] = False
                eval_record["reasoning"].append(f"Hard-coded check: {numbers} CANNOT reach 24 with any single operation")
                eval_record["final_value"] = 0.001
                eval_record["llm_judgments"] = ["impossible", "impossible", "impossible"]
                eval_record["from_cache"] = False
                eval_record["score_breakdown"] = {
                    "sure_count": 0,
                    "likely_count": 0,
                    "impossible_count": 3,
                    "n_evaluate_sample": 3,
                    "raw_score": 0.003,
                    "boosted_score": 0.003,
                    "confidence_level": "impossible"
                }
                return 0.001, eval_record  # Dead end!
        
        # === LLM EVALUATION (For nuanced cases with 3+ numbers) ===
        
        numbers_str = str(numbers)
        
        # Use appropriate prompt
        if is_final:
            # For final validation, would need the expression
            prompt = VALUE_PROMPT_CODEACT.format(input=numbers_str)
        else:
            prompt = VALUE_PROMPT_CODEACT.format(input=numbers_str)
        
        eval_record["llm_prompt"] = prompt
        
        # Check cache
        if prompt in self.value_cache:
            self.stats['cache_hits'] += 1
            # Return cached value with minimal eval_record (cache hit)
            cached_value = self.value_cache[prompt]
            eval_record["from_cache"] = True
            eval_record["final_value"] = cached_value
            eval_record["reasoning"].append("Retrieved from cache (previously evaluated)")
            return cached_value, eval_record
        
        eval_record["from_cache"] = False
        
        # Check rate limits before API calls
        self.check_rate_limits()
        
        # Get evaluations with rate limiting
        time.sleep(self.api_delay)
        self.stats['api_calls'] += self.n_evaluate_sample
        self.stats['daily_requests'] += self.n_evaluate_sample
        
        value_outputs = []
        for _ in range(self.n_evaluate_sample):
            try:
                response = openai_generate(prompt, n=1, temperature=0.0, system_prompt=SYSTEM_PROMPT)[0]  # Deterministic evaluation
                value_outputs.append(response.strip().lower())
            except:
                value_outputs.append("likely")
        
        # Parse evaluations
        value_map = {'impossible': 0.001, 'likely': 1, 'sure': 20}
        value_names = []
        
        for output in value_outputs:
            # Extract last word
            last_word = output.strip().split()[-1] if output.strip() else "likely"
            value_names.append(last_word)
        
        eval_record["llm_judgments"] = value_names
        eval_record["llm_raw_responses"] = value_outputs
        
        # Count judgments
        sure_count = value_names.count("sure")
        likely_count = value_names.count("likely")
        impossible_count = value_names.count("impossible")
        
        eval_record["score_breakdown"] = {
            "sure_count": sure_count,
            "likely_count": likely_count,
            "impossible_count": impossible_count,
            "n_evaluate_sample": self.n_evaluate_sample
        }
        
        # Calculate score
        raw_value = sum(value_map.get(name, 1) * value_names.count(name) for name in value_map.keys())
        
        eval_record["score_breakdown"]["raw_score"] = raw_value
        
        # Apply boost for promising intermediate states
        boosted_value = raw_value * llm_boost
        
        eval_record["score_breakdown"]["boosted_score"] = boosted_value
        
        # Generate reasoning based on LLM judgments
        if sure_count >= 2:
            confidence = "high"
            eval_record["reasoning"].append(f"LLM is confident this can reach 24 (judgments: {value_names})")
        elif impossible_count >= 2:
            confidence = "very_low"
            eval_record["reasoning"].append(f"LLM judges this path as impossible (judgments: {value_names})")
        elif likely_count == self.n_evaluate_sample:
            confidence = "medium"
            eval_record["reasoning"].append(f"LLM sees potential but is uncertain (judgments: {value_names})")
        else:
            confidence = "mixed"
            eval_record["reasoning"].append(f"LLM has mixed opinions (judgments: {value_names})")
        
        eval_record["score_breakdown"]["confidence_level"] = confidence
        eval_record["final_value"] = boosted_value
        
        # Cache result (cache only the numeric value)
        self.value_cache[prompt] = boosted_value
        
        return boosted_value, eval_record
    
    def solve(self, numbers: List[int], verbose: bool = True) -> Tuple[List[str], TreeNode]:
        """
        Solve Game of 24 using Tree of Thoughts with CodeAct
        
        Args:
            numbers: List of 4 numbers
            verbose: Print progress
        
        Returns:
            Tuple of (solutions, root_node)
        """
        # Reset for new problem
        TreeNode.node_counter = 0
        self.all_nodes = []
        self.solutions = []
        
        # Create root node
        self.root = TreeNode(
            state="", 
            parent=None, 
            action="START", 
            value=0.0,
            thought="Initial state",
            code="",
            observation=f"Starting numbers: {numbers}",
            path_history=""  # Empty for root
        )
        self.all_nodes.append(self.root)
        
        # Track current numbers for each node
        node_to_numbers = {self.root: numbers}
        
        # Track all seen states globally to avoid revisiting
        global_seen_states = set()
        global_seen_states.add(tuple(sorted(numbers)))
        
        # BFS search
        current_nodes = [self.root]
        
        for step in range(self.max_steps):
            if verbose:
                print(f"\n{'='*70}")
                print(f"STEP {step + 1}/{self.max_steps}")
                print(f"Current candidates: {len(current_nodes)}")
            
            new_nodes = []
            
            # === SPECIAL HANDLING FOR EXHAUSTIVE DEPTH-1 MODE ===
            if step == 0 and self.exhaustive_depth1:
                # At depth 1, generate ALL possible first moves (no LLM)
                if verbose:
                    print(f"\n  🔬 EXHAUSTIVE DEPTH-1 MODE ACTIVATED")
                    print(f"  Generating ALL ~24 possible first moves instead of LLM proposals...")
                
                # Only one node at step 0 (the root)
                node = current_nodes[0]
                current_nums = node_to_numbers[node]
                
                # Generate ALL possible first moves exhaustively
                proposals = self.generate_all_first_moves(current_nums)
                
                if verbose:
                    print(f"  ✓ Generated {len(proposals)} exhaustive first moves (0 API calls)")
                
                # Create child nodes for ALL proposals
                for prop in proposals:
                    new_state_str = str(prop['new_state'])
                    state_tuple = tuple(sorted(prop['new_state']))
                    
                    # Skip if duplicate (shouldn't happen with exhaustive generation, but safety check)
                    if state_tuple in global_seen_states:
                        continue
                    
                    global_seen_states.add(state_tuple)
                    
                    # Build path history
                    step_description = f"{prop['thought']}\nCode: {prop['code']}\nResult: {prop['observation']}"
                    new_path_history = step_description  # First step, no parent history
                    
                    child = TreeNode(
                        state=new_state_str,
                        parent=node,
                        action=prop['action'],
                        value=0.0,  # Will be set during evaluation
                        thought=prop['thought'],
                        code=prop['code'],
                        observation=prop['observation'],
                        path_history=new_path_history
                    )
                    new_nodes.append(child)
                    self.all_nodes.append(child)
                    node_to_numbers[child] = prop['new_state']
            
            # === NORMAL LLM PROPOSALS MODE (for depth 2+ or when exhaustive_depth1=False) ===
            else:
                # Generate proposals for each current state
                for node in current_nodes:
                    current_nums = node_to_numbers[node]
                    
                    if len(current_nums) == 1:
                        # Already reached single number
                        continue
                    
                    if verbose:
                        print(f"\n  Node {node.id}: Generating proposals for {current_nums}")
                    
                    # Collect states from this node's children (already tried from this node)
                    avoid_for_this_node = []
                    for child in node.children:
                        if child in node_to_numbers:
                            avoid_for_this_node.append(node_to_numbers[child])
                    
                    # Generate multiple proposals (increase for diversity)
                    # Pass avoid_states to prevent duplicates within this node
                    # Also pass original_input and path_history for context (like original ToT paper)
                    proposals = self.get_proposals(
                        current_nums, 
                        original_input=numbers,  # Original puzzle for context
                        path_history=node.path_history,  # Full history of steps
                        n_proposals=5, 
                        avoid_states=avoid_for_this_node
                    )
                    
                    if verbose:
                        print(f"    → Generated {len(proposals)} unique proposals")
                    
                    # Create child nodes (only if state not globally seen)
                    for prop in proposals:
                        new_state_str = str(prop['new_state'])
                        state_tuple = tuple(sorted(prop['new_state']))
                        
                        # Skip if we've seen this exact state before anywhere in the tree
                        if state_tuple in global_seen_states:
                            if verbose:
                                print(f"    ⊗ Skipping duplicate state: {prop['new_state']}")
                            continue
                        
                        global_seen_states.add(state_tuple)
                    
                    # Create child nodes
                    for prop in proposals:
                        new_state_str = str(prop['new_state'])
                        
                        # Build path history (like original ToT paper)
                        # Accumulate: parent_history + current_step
                        step_description = f"{prop['thought']}\nCode: {prop['code']}\nResult: {prop['observation']}"
                        new_path_history = node.path_history + "\n" + step_description if node.path_history else step_description
                        
                        child = TreeNode(
                            state=new_state_str,
                            parent=node,
                            action=prop['action'],
                            value=0.0,  # Will be set during evaluation
                            thought=prop['thought'],
                            code=prop['code'],
                            observation=prop['observation'],
                            path_history=new_path_history  # Pass accumulated history
                        )
                        new_nodes.append(child)
                        self.all_nodes.append(child)
                        node_to_numbers[child] = prop['new_state']
            
            if not new_nodes:
                if verbose:
                    print("\n  ⚠ No new proposals generated. Stopping.")
                break
            
            # Evaluate all new nodes
            if verbose:
                print(f"\n  Evaluating {len(new_nodes)} new states...")
            
            for node in new_nodes:
                nums = node_to_numbers[node]
                value, eval_record = self.evaluate_state(nums, is_final=(len(nums) == 1))
                node.value = value
                node.evaluation = eval_record  # Store detailed evaluation for distillation
            
            # CRITICAL: Prune nodes with premature 24 (don't expand them at all)
            # EXCEPTION: [24, 1] or [24, 1, 1] is OK because 24*1=24 or 24/1=24
            viable_nodes = []
            pruned_trap_nodes = []
            for node in new_nodes:
                nums = node_to_numbers[node]
                # If we have multiple numbers and 24 is one of them, check if it's a trap
                if len(nums) > 1 and (24 in nums or 24.0 in nums):
                    # Get numbers other than 24
                    other_nums = [x for x in nums if abs(x - 24) > 0.001]
                    # Only prune if there are non-1 numbers remaining
                    # [24, 1] or [24, 1, 1] is OK! Can multiply/divide by 1 to get 24
                    if other_nums and not all(abs(x - 1) < 0.001 for x in other_nums):
                        node.is_pruned = True
                        pruned_trap_nodes.append(node)
                        if verbose:
                            print(f"    ✂ PRUNED TRAP: {nums} (premature 24 with non-1 numbers)")
                    else:
                        # [24, 1] is valid - keep it!
                        viable_nodes.append(node)
                        if verbose and other_nums:
                            print(f"    ✓ KEPT: {nums} (24 with only 1s - can reach solution)")
                else:
                    viable_nodes.append(node)
            
            if verbose and pruned_trap_nodes:
                print(f"  ✂ Pruned {len(pruned_trap_nodes)} trap nodes (premature 24)")
            
            # === SELECTIVE EXHAUSTIVE RESCUE ===
            # If at depth-1 and all LLM proposals score poorly, regenerate exhaustively
            # Controlled by enable_ser hyperparameter
            if self.enable_ser and step == 0 and not self.exhaustive_depth1:
                # Check if LLM proposals are all low-confidence
                if viable_nodes:
                    max_value = max(node.value for node in viable_nodes)
                    LOW_CONFIDENCE_THRESHOLD = 5.0  # Trigger only if NO "sure" ratings (20.0) in any proposal
                    # Threshold rationale:
                    # - "likely, likely, sure" = 1+1+20 = 22.0 (above threshold, no SER)
                    # - "likely, likely, likely" = 1+1+1 = 3.0 (below threshold, SER triggers)
                    # - This makes SER selective: triggers only when LLM finds NO promising paths
                    
                    if max_value < LOW_CONFIDENCE_THRESHOLD:
                        if verbose:
                            print(f"\n  ⚠️ SELECTIVE EXHAUSTIVE RESCUE TRIGGERED!")
                            print(f"  → Max depth-1 value: {max_value:.2f} < threshold {LOW_CONFIDENCE_THRESHOLD}")
                            print(f"  → LLM proposals seem weak, regenerating ALL first moves exhaustively...")
                        
                        # Mark existing LLM proposals as pruned (will be replaced)
                        for node in viable_nodes:
                            node.is_pruned = True
                        
                        # Generate ALL possible first moves exhaustively
                        node = current_nodes[0]  # Root node
                        current_nums = node_to_numbers[node]
                        proposals = self.generate_all_first_moves(current_nums)
                        
                        if verbose:
                            print(f"  ✓ Generated {len(proposals)} exhaustive first moves")
                        
                        # Create new child nodes for ALL proposals
                        exhaustive_nodes = []
                        for prop in proposals:
                            new_state_str = str(prop['new_state'])
                            state_tuple = tuple(sorted(prop['new_state']))
                            
                            # Skip if duplicate
                            if state_tuple in global_seen_states:
                                continue
                            
                            global_seen_states.add(state_tuple)
                            
                            # Build path history
                            step_description = f"{prop['thought']}\nCode: {prop['code']}\nResult: {prop['observation']}"
                            new_path_history = step_description
                            
                            child = TreeNode(
                                state=new_state_str,
                                parent=node,
                                action=prop['action'],
                                value=0.0,
                                thought=prop['thought'],
                                code=prop['code'],
                                observation=prop['observation'],
                                path_history=new_path_history
                            )
                            exhaustive_nodes.append(child)
                            self.all_nodes.append(child)
                            node_to_numbers[child] = prop['new_state']
                        
                        # Evaluate exhaustive nodes (use single sample for speed)
                        if verbose:
                            print(f"  Evaluating {len(exhaustive_nodes)} exhaustive proposals (n=1 for speed)...")
                        
                        # Temporarily reduce evaluation samples for exhaustive rescue
                        original_n_eval = self.n_evaluate_sample
                        self.n_evaluate_sample = 1  # Single evaluation is enough for exhaustive moves
                        
                        for node in exhaustive_nodes:
                            nums = node_to_numbers[node]
                            value, eval_record = self.evaluate_state(nums, is_final=(len(nums) == 1))
                            node.value = value
                            node.evaluation = eval_record
                        
                        # Restore original setting
                        self.n_evaluate_sample = original_n_eval
                        
                        # Replace viable_nodes with exhaustive_nodes
                        viable_nodes = exhaustive_nodes
                        
                        # Check for premature 24 trap in exhaustive nodes
                        pruned_trap_nodes = []
                        viable_after_trap_check = []
                        for node in viable_nodes:
                            nums = node_to_numbers[node]
                            if len(nums) > 1 and (24 in nums or 24.0 in nums):
                                other_nums = [x for x in nums if abs(x - 24) > 0.001]
                                if other_nums and not all(abs(x - 1) < 0.001 for x in other_nums):
                                    node.is_pruned = True
                                    pruned_trap_nodes.append(node)
                                else:
                                    viable_after_trap_check.append(node)
                            else:
                                viable_after_trap_check.append(node)
                        
                        viable_nodes = viable_after_trap_check
                        
                        if verbose:
                            print(f"  🔬 Exhaustive rescue complete: {len(viable_nodes)} viable states")
                            if pruned_trap_nodes:
                                print(f"  ✂ Pruned {len(pruned_trap_nodes)} premature 24 traps")
            
            # Select nodes using chosen method (greedy or probabilistic)
            viable_nodes.sort(key=lambda x: x.value, reverse=True)
            
            if self.selection_method == 'greedy':
                # Original: Select top-k by value (deterministic)
                selected_nodes = viable_nodes[:self.n_select_sample]
            elif self.selection_method == 'sample':
                # Probabilistic: Sample based on values (more exploration)
                # From original ToT paper: ps = values / sum(values)
                if len(viable_nodes) <= self.n_select_sample:
                    selected_nodes = viable_nodes
                else:
                    import numpy as np
                    values = np.array([node.value for node in viable_nodes])
                    if values.sum() > 0:
                        probs = values / values.sum()
                        indices = np.random.choice(
                            len(viable_nodes), 
                            size=min(self.n_select_sample, len(viable_nodes)),
                            p=probs,
                            replace=False
                        )
                        selected_nodes = [viable_nodes[i] for i in indices]
                    else:
                        # Fallback to greedy if all values are 0
                        selected_nodes = viable_nodes[:self.n_select_sample]
            else:
                # Default to greedy
                selected_nodes = viable_nodes[:self.n_select_sample]
            
            # 🔧 DELAYED FRACTION PRESERVATION (DFP)
            # Before pruning, rescue fragile fractional states
            dfp_rescued = []
            for node in viable_nodes:
                if node not in selected_nodes:
                    nums = node_to_numbers[node]
                    non_int_count, small_int_count = state_signature(nums)
                    
                    # DFP Rule: 1 non-integer + at least 1 small integer (≤6)
                    # This catches states like [0.571, 8, 10] which need preservation
                    if non_int_count == 1 and small_int_count >= 1:
                        selected_nodes.append(node)
                        dfp_rescued.append(node)
                        if verbose:
                            print(f"    🔧 DFP RESCUED: {nums} (fragile fraction preserved)")
            
            if verbose and dfp_rescued:
                print(f"  🔧 DFP rescued {len(dfp_rescued)} fragile fractional states")
            
            # Mark remaining pruned nodes
            for node in viable_nodes:
                if node not in selected_nodes:
                    node.is_pruned = True
            
            if verbose:
                method_name = "🎯 Greedy" if self.selection_method == 'greedy' else "🎲 Probabilistic"
                print(f"\n  {method_name} top {len(selected_nodes)} selected:")
                for i, node in enumerate(selected_nodes[:3]):
                    nums = node_to_numbers[node]
                    print(f"    {i+1}. Value={node.value:.2f} | Numbers={nums}")
                    print(f"       Thought: {node.thought[:60]}...")
                    print(f"       Thought: {node.thought[:60]}...")
            
            current_nodes = selected_nodes
        
        # Check for solutions
        if verbose:
            print(f"\n{'='*70}")
            print("CHECKING SOLUTIONS...")
        
        for node in self.all_nodes:
            if node in node_to_numbers:
                nums = node_to_numbers[node]
                # Use tolerance for floating-point comparison (handles 23.999999... = 24)
                if len(nums) == 1 and abs(nums[0] - 24) < 0.001:
                    node.is_solution = True
                    self.solutions.append(node)
        
        self.stats['total_nodes'] = len(self.all_nodes)
        self.stats['solutions_found'] = len(self.solutions)
        
        if verbose:
            print(f"\n✓ Found {len(self.solutions)} valid solution(s)")
            print(f"\nStatistics:")
            print(f"  Total nodes explored: {self.stats['total_nodes']}")
            print(f"  API calls made: {self.stats['api_calls']}")
            print(f"  Cache hits: {self.stats['cache_hits']}")
            print(f"  Code executions: {self.stats['code_executions']}")
            print(f"  Code errors: {self.stats['code_errors']}")
        
        return [self.reconstruct_solution_path(node) for node in self.solutions], self.root
    
    def reconstruct_solution_path(self, node: TreeNode) -> str:
        """Reconstruct the solution path from root to node"""
        path = []
        current = node
        
        while current.parent is not None:
            step_info = f"Thought: {current.thought}\nCode:\n{current.code}\nResult: {current.observation}"
            path.append(step_info)
            current = current.parent
        
        path.reverse()
        return "\n\n".join(path)
    
    def export_tree_to_json(self, filename: str = None) -> str:
        """
        Export the entire search tree to JSON
        
        Args:
            filename: Output filename (default: auto-generated)
        
        Returns:
            Filename where tree was saved
        """
        if filename is None:
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
            filename = f"game24_codeact_tree_{timestamp}.json"
        
        # Convert stats to JSON-serializable format (convert datetime to string)
        stats_serializable = self.stats.copy()
        if 'session_start' in stats_serializable:
            try:
                stats_serializable['session_start'] = stats_serializable['session_start'].isoformat()
            except AttributeError:
                pass  # Already a string or not datetime
        
        tree_data = {
            'metadata': {
                'timestamp': datetime.now().isoformat(),
                'mode': 'CodeAct',
                'parameters': {
                    'temperature': self.temperature,
                    'n_evaluate_sample': self.n_evaluate_sample,
                    'n_select_sample': self.n_select_sample,
                    'max_steps': self.max_steps,
                    'api_delay': self.api_delay
                },
                'statistics': stats_serializable
            },
            'nodes': [node.to_dict() for node in self.all_nodes],
            'solutions': [node.id for node in self.solutions]
        }
        
        with open(filename, 'w', encoding='utf-8') as f:
            json.dump(tree_data, f, indent=2, ensure_ascii=False)
        
        print(f"✓ Tree exported to: {filename}")
        return filename

print("✓ Game24TreeOfThoughts with CodeAct class defined")
````

# Game of 24 - Tree of Thoughts Solver with CodeAct 🎮🌳💻

## Overview
This notebook implements a **Tree of Thoughts (ToT)** solver for the Game of 24 puzzle using **CodeAct pattern** with Google's **Gemini API**.

### What is Game of 24?
Given 4 numbers, use basic arithmetic operations (+, -, *, /) to make 24. Each number must be used exactly once.

**Example:** `[4, 5, 6, 10]`
- Solution: `(10 - 6) * (4 + 5) = 24`

---

## 🚀 What is CodeAct?

**CodeAct** is an advanced agent pattern where instead of directly generating answers, the LLM:

1. **Thinks** about what to do next (Thought)
2. **Writes executable Python code** to perform the action (Code)
3. **Observes** the result from running the code (Observation)

### Why CodeAct > Direct Generation?

| Direct LLM | CodeAct Pattern |
|------------|-----------------|
| LLM does mental math (error-prone) | Python does calculation (exact) |
| No verification | Executable = verifiable |
| Hallucination risk | Grounded in code execution |
| Black box | Transparent reasoning |

---

## How It Works

### Traditional ToT Flow:
```
Generate → Evaluate → Select → Repeat
```

### CodeAct ToT Flow (This Implementation):
```
For each node:
  1. THOUGHT: "I'll add numbers[0] and numbers[1]"
  2. CODE:    "res = numbers[0] + numbers[1]; ..."
  3. EXECUTE: Run code in safe sandbox
  4. OBSERVE: "[9, 6, 10]"
  5. STORE:   Save in tree node
  
Then: Evaluate → Select → Repeat
```

---

## 🔒 Safe Sandbox

Code is executed in a **restricted environment**:

### ✅ Allowed:
- Basic math operations
- List operations
- `math` module
- Safe built-ins (print, len, sum, etc.)

### ❌ Blocked:
- File I/O (`open`, file operations)
- Network access
- OS commands (`os`, `sys`, `subprocess`)
- Dangerous functions (`exec`, `eval`, `__import__`)

**Result:** LLM can generate code freely, but it's safely sandboxed!

---

Each node stores:

## 🎯 Understanding Beam Search Pruning (n_select_sample)

### Key Concept: GLOBAL Pruning Across Entire Depth Level

**`n_select_sample` is NOT per-node, it's per-depth:**

```
Depth 1: 5 parent nodes
    ↓ (each generates ~5 proposals)
    
Depth 2: 25 total children created (5 × 5)
    ↓
GLOBAL PRUNING: Evaluate ALL 25 children together
    ↓
Keep top n_select_sample=10 BEST across ALL parents
    ↓
Result: 10 nodes at depth 2 (not 5×10=50!)
```

### Example Scenario:

**Before Pruning (25 children from 5 parents):**
- Parent A contributes: 5 children (values: 8.5, 7.2, 6.1, 5.3, 4.8)
- Parent B contributes: 5 children (values: 9.1, 8.8, 8.3, 7.9, 7.5)
- Parent C contributes: 5 children (values: 3.2, 2.8, 2.1, 1.9, 1.5)
- Parent D contributes: 5 children (values: 6.8, 6.5, 6.2, 5.9, 5.6)
- Parent E contributes: 5 children (values: 2.5, 2.3, 2.0, 1.8, 1.6)

**After Pruning (keep top 10 globally):**
- From Parent A: 2 children kept (8.5, 7.2)
- From Parent B: **5 children kept** (9.1, 8.8, 8.3, 7.9, 7.5) ← ALL kept!
- From Parent C: 0 children kept ← ALL pruned!
- From Parent D: 3 children kept (6.8, 6.5, 6.2)
- From Parent E: 0 children kept ← ALL pruned!

**Result:** Some parents are "dead ends" (all children pruned), others are "promising paths" (multiple children kept).

### Why This Matters:

1. **Explores promising branches deeper**: If Parent B has great children, we explore all of them
2. **Abandons dead-end branches**: If Parent C/E have poor children, we stop exploring them
3. **Dynamic resource allocation**: Beam width adapts to where the good solutions are

### Special Cases:

- **Root → Depth 1**: No pruning (LLM generates 5-8, all kept since < beam width)
- **Depth 1 → Depth 2**: First pruning happens here (5×5=25 → keep top 10)
- **Later depths**: Continue global pruning (keep top 10 across all parents)

## 🔬 Selective Exhaustive Rescue (SER)

### The Problem: LLM Proposal Bias

**Observation:** When puzzles require ONE critical first move, GPT-4o often misses it:
- `[2,4,8,9]`: Needs `4+8=12` → Never proposed
- `[6,7,8,9]`: Needs `9-7=2` → Never proposed  
- `[2,3,5,12]`: Needs `5/2=2.5` → Never proposed

**Root cause:** LLMs avoid unconventional moves (small subtractions, fractions, "useless" sums)

---

### The Solution: Adaptive Exhaustive Search

**Strategy:** Only use exhaustive search when LLM proposals are clearly weak.

**Logic:**
```python
if step == 0:  # After first depth
    if max(node.value) < LOW_CONFIDENCE_THRESHOLD:
        # LLM proposals are all weak → regenerate exhaustively
        proposals = generate_all_first_moves(numbers)
```

**Threshold:** `5.0` (below "likely, likely" with boost)

---

### Why This Works

**Cost-efficient:**
- ✅ Most puzzles: Pure LLM (0 extra cost)
- ⚠️ Hard puzzles: Triggers rescue (~24 exhaustive first moves)
- Only adds cost when needed!

**Quality:**
- ✅ Solves single-path puzzles ([2,4,8,9], [6,7,8,9])
- ✅ Maintains LLM reasoning for depths 2+ (good for distillation)
- ✅ Transparent: Logs when rescue triggers

**Trade-offs:**
- ✅ Better than always-exhaustive (more principled)
- ✅ Better than pure LLM (higher success rate)
- ⚠️ Still hybrid approach (not 100% LLM)

---

### Test Cases

**Easy puzzles** → LLM succeeds, no rescue needed  
**Hard puzzles** → LLM fails depth-1, rescue activates, finds solution

````python
def test_ser_trigger_fixed(puzzle, description):
    """Test if SER would trigger for a given puzzle - FIXED VERSION"""
    test_solver = Game24TreeOfThoughts(
        n_evaluate_sample=3,
        n_select_sample=10,
        max_steps=6,
        exhaustive_depth1=False
    )
    
    # Generate proposals - CORRECT WAY
    proposals = test_solver.get_proposals(
        current_numbers=puzzle,
        original_input=puzzle,
        path_history="",
        n_proposals=5,  # Ask for 5 proposals
        avoid_states=[]
    )
    
    # Evaluate each proposal
    proposal_values = []
    proposal_states = []
    for prop in proposals:
        new_state = prop['new_state']  # Already a list, don't need eval()
        value, _ = test_solver.evaluate_state(
            new_state, 
            is_final=(len(new_state) == 1)
        )
        proposal_values.append(value)
        proposal_states.append(str(new_state))
    
    # Calculate trigger decision
    max_val = max(proposal_values) if proposal_values else 0
    threshold = 10.0
    triggers = max_val < threshold
    
    return {
        'Puzzle': str(puzzle),
        'Description': description,
        'Num Proposals': len(proposals),
        'Max Value': round(max_val, 1),
        'SER Triggers?': '🚨 YES' if triggers else '✅ NO',
        'Top Proposals': ', '.join(proposal_states[:3]) if proposal_states else 'None'
    }

print("✅ Fixed test function defined!")
````
